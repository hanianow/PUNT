Automatic access and printing labels from custom Harvard Oxford atlas. It is a custom one as I renumbered one of the hemispheres to bigger numbers but if needed you can adapt the code to use normally 0-48 labels.
https://github.com/hanianow , July 2018

```python
import untangle

atlas_obj = untangle.parse('/usr/local/fsl/data/atlases/HarvardOxford-Cortical.xml') # change to your own FSL path
```

```python
#check if it reads atlas correctly:
x = atlas_obj.atlas.data.label[37]
x.cdata 
```

I'm iterating here over two endings of two WM tracts but of course it can be change to anything that describes txt files

```python
def print_regions(ending, tractname):
    tractns = open( "../" + ending + "_" + tractname + "_for_matlab", "r" )
    for line in tractns:
        n=int(line)
        if n==101:
            region = 'Left Thalamus'
        elif n==102:
            region = 'Right Thalamus'
        elif n <=48:
            nlabel = atlas_obj.atlas.data.label[n-1]
            region = nlabel.cdata
        elif n>48:
            nlabel = atlas_obj.atlas.data.label[n-48-1]   
            region = nlabel.cdata
        print(region)
```

```python
tracts = ['ifo_l','ifo_r','slf_l','slf_r','ilf_l','ilf_r','unc_l','unc_r','fma','fmi', 'ar_l', 'ar_r','atr_l','atr_r', 'cgc_l', 'cgc_r', 'cgh_l', 'cgh_r','ptr_l', 'ptr_r', 'str_l', 'str_r']
for t in tracts:
    print(t)
    print('Ending A:')
    print_regions('A', t)
    print('Ending B:')
    print_regions('B', t)
    print('............')
```
